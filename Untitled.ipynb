```python
import os
from PIL import Image

def convert_bulk(path: str, new_dir: str):
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
    
    for image_name in os.listdir(path):
        image_path = os.path.join(path, image_name)
        try :
            with Image.open(image_path) as img:
                conv_image = img.convert('RGB')
                conv_image.thumbnail((400, 400))
                new_image_path = os.path.join(new_dir, os.path.splitext(image_name)[0] + ".png")
                conv_image.save(new_image_path, "PNG")
        except Exception as e:
            print(e)
            continue
```

```python
convert_bulk('images/images/', 'new_images')
```

```python
from datasets import load_dataset

data = load_dataset("new_images", split="train")
data
```

```python
from sentence_transformers import SentenceTransformer

images = data['image']
```

```python
model = SentenceTransformer('clip-ViT-B-32')
image_embeddings = model.encode([image for image in images])
```

```python
from qdrant_client import QdrantClient
from qdrant_client.http import models

# Initialize Qdrant client and create a collection
client = QdrantClient(url='http://localhost:6333',api_key='helloworld')
client.recreate_collection(
    collection_name="mycollection2",
    vectors_config={"image": models.VectorParams(size=512, distance=models.Distance.COSINE)}
)
```

```python
from PIL import Image
from io import BytesIO
import base64

def pil_image_to_base64(pil_image):
    """
    Convert a PIL image to a base64-encoded string.

    Args:
        pil_image (PIL.JpegImagePlugin.JpegImageFile): A PIL image object.

    Returns:
        str: The base64-encoded image data.
    """
    byte_io = BytesIO()
    pil_image.save(byte_io, format='PNG')
    byte_image = byte_io.getvalue()
    base64_image = base64.b64encode(byte_image).decode('utf-8')
    return base64_image
```

```python
import uuid
from qdrant_client.http.models import PointStruct
from tqdm import tqdm

points = []

for embedding , image in tqdm(zip(image_embeddings , images)) :
    payload = {"base64_image": pil_image_to_base64(image)}
    embeddings = embedding.tolist()
    point_id = str(uuid.uuid4())  # Generate a unique ID for the point
    points.append(PointStruct(id=point_id, payload=payload, vector={"image": embeddings}))

```

```python
batch_size = 100
for i in tqdm(range(0, len(points), batch_size)):
    batch = points[i:i + batch_size]
    client.upsert(
        collection_name="mycollection2",
        wait=True,
        points=batch
    )
```

```python
from IPython.display import display
import io
def display_base64_image(base64_string):
    image_data = base64.b64decode(base64_string)
    image_bytes = io.BytesIO(image_data)
    image = Image.open(image_bytes)
    display(image)
```
